## Advance ORM Query 

### การซื้อครั้งไหนบ้างที่มีสินค้าราคามากกว่า $50 รวมอยู่ด้วย

```python
# การซื้อครั้งไหนบ้างที่มีสินค้าราคามากกว่า $50 รวมอยู่ด้วย
query = Transaction.objects.filter(transactionitem__item__price__gt=50).values()

df = pd.DataFrame(list( query ))
df
```

### ลูกค้าคนไหนบ้างที่เพิ่มสินค้าไปในตะกร้าแต่ยังไม่ตัดสินใจซื้อสินค้า

```python
# ลูกค้าคนไหนบ้างที่เพิ่มสินค้าไปในตะกร้าแต่ยังไม่ตัดสินใจซื้อสินค้า
query = Customer.objects.filter(transaction__complete=False).values()

df = pd.DataFrame(list( query ))
df
```

### ลูกค้าคนไหนบ้างที่ซื้อสินค้าแล้ว แต่ยังไม่ได้แนบหลักฐานการโอนเงิน

```python
# ลูกค้าคนไหนบ้างที่ซื้อสินค้าแล้ว แต่ยังไม่ได้แนบหลักฐานการโอนเงิน
query = Customer.objects.filter(transaction__evidence='').values().distinct()

df = pd.DataFrame(list( query ))
df
```
